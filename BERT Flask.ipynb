```
from flask import Flask, render_template, request
from flask_ngrok import run_with_ngrok
#!python3
#orca
import ktrain
app = Flask(__name__,template_folder='/content/drive/My Drive/Other Projects/flask/bert sentiment/templates')
run_with_ngrok(app)  # Start ngrok when app is run

@app.route('/', methods=['GET','POST'])
def index():
    op_sent=''
    predictor = ktrain.load_predictor('/content/drive/My Drive/Other Projects/flask/bert sentiment/sentiment_predictor')    
    if request.method=='POST' and 'sentiment' in request.form:
        sentiment = request.form.get('sentiment')
        op_sent = sentiment_analyse(sentiment)
    return render_template("index.html",op_sent=op_sent)

def sentiment_analyse(sentiment):
    return predictor.predict(sentiment)[0]
    
app.run()    
```

```
tester = ["I think this is awesome!"]
if predictor.predict(tester)[0]=='pos':
  print('positive')
else:
  print('negative')
```
